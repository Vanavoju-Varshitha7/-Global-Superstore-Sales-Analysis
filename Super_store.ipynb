```python
import pandas as pd
import numpy as np
df= pd.read_csv('SuperStore_Orders.csv', encoding='latin1')
print(df.columns.to_list())
y=df.columns.to_list()
y
```

```python
print("The shape of the dataframe:",np.shape(df))
print("-------------------------------------------------------------------------")
print("the first five rows are:\n",df.head())
print("-------------------------------------------------------------------------")
for i in y:
    print("the dtype of",i,"is:",df[i].dtype)
print("-------------------------------------------------------------------------")
print("the null count of each column is :\n",df.isnull().sum())
```

```python
print("is any null values exist:",df.isnull())
```

```python
#drop the duplicates
df.drop_duplicates()
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format='%d-%m-%Y')
df['ship_date']=pd.to_datetime(df['ship_date'],format='%d-%m-%Y')
np.shape(df)
```

```python
print(df.duplicated().sum())  #if we want to know any duplicates are there then we use this
```

```python
df['sales']=pd.to_numeric(df['sales'],errors='coerce')
Total_sales=(np.nansum(df['sales'].values).round(2))
print("the total sales across all orders:",Total_sales)
df['profit']=pd.to_numeric(df['profit'],errors='coerce')
Total_profits=(np.nansum(df['profit'].values).round(2))
print("the total profits across all orders:",Total_profits)
Total_quantity=(np.sum(df['quantity'].values))
print("the total quantity across all orders:",Total_quantity)
Total_shipping_cost=(np.sum(df['shipping_cost'].values).round(2))
print("the total shiping cost across all orders:",Total_shipping_cost)
Overall_profit_margin = ((Total_profits / Total_sales) *100).round(2)
print("Overall profits margin are:",Overall_profit_margin)
```

```python
# Find total sales per category
print("The total sales per category are:\n",df.groupby('category')['sales'].sum())
print("-------------------------------------------------------------------------")
print("The total profit per category are:\n",df.groupby('category')['profit'].sum())
print("-------------------------------------------------------------------------")
print("Sort by sales descending:")
df.sort_values(by='sales',ascending=False)
```

```python
# Group by state
print("The total sales per state are:\n",df.groupby('state')['sales'].sum())
print("-------------------------------------------------------------------------")
print("The total profit per state are:\n",df.groupby('state')['profit'].sum())
print("-------------------------------------------------------------------------")
```

```python
df.sort_values(by='sales',ascending=False)
print("The top 10 states are:\n",df.head(10))
print("-------------------------------------------------------------------------")
print("The bottom 5 states are:\n",df.tail())
```

```python
df['profit'].dtype
```

```python
print("the mean of the profit per order:",np.mean(df['profit']).round(2))  #df['profit'].mean().round(2)
print("the median of the profit per order:",np.median(df['profit']))
print("the std of the profit per order:",(round(np.std(df['profit']),(2))))
print("the max of profit per order:",np.max(df['profit']))
print("the min of profit per order:",np.min(df['profit']))
print("-------------------------------------------------------------------------")
count_loss=len(df['profit'][df['profit']<0].values)  #if i want to get the count of the rows after the condition is applies ,here i want how many negative profits,so first negative profits ,convert into the array and find the length of that array by using len
# len(y['profit'].values)
print("the orders which have negative orders are:",count_loss)
loss_percentage=((count_loss/len(df['profit'].values))*100)
print(f'the loss percentage is:{loss_percentage:.2f}')
# Print the top 5 loss-making orders showing product_name, sales, profit)
top_5=df[['product_name','sales','profit']][df['profit']<0].sort_values(by='profit').head()
top_5
```

```python
print("=== SEGMENT ANALYSIS ===")
seg = df.groupby('segment').agg(
    total_sales   = ('sales',    'sum'),
    total_profit  = ('profit',   'sum'),
    avg_discount  = ('discount', 'mean')
).round(2)

seg['profit_margin'] = np.round((seg['total_profit'] / seg['total_sales']) * 100, 2)
print(seg)

best_seg = seg['profit_margin'].idxmax()
print(f"\nHighest profit margin segment: {best_seg}")



```

```python
print("=== DISCOUNT IMPACT ON PROFIT ===")
corr = np.corrcoef(df['discount'].values, df['profit'].values)[0,1]
print(f"Correlation (Discount vs Profit): {corr:.4f}")

if corr < -0.3:
    print("Negative correlation — higher discount = lower profit")
elif corr < 0:
    print("Slight negative correlation — discount slightly hurts profit")
else:
    print("Positive correlation — discount does not hurt profit")

high_disc = df[df['discount'] >  0.2]['profit'].mean()
low_disc  = df[df['discount'] <= 0.2]['profit'].mean()
print(f"\nAvg profit with HIGH discount (>20%) : ${high_disc:,.2f}")
print(f"Avg profit with LOW  discount (<=20%): ${low_disc:,.2f}")
```

```python
# Group by year and find total sales and total profit
group_by_year=df.groupby('year').agg(
    total_sales=('sales','sum'),
    total_profits=('profit','sum')
    ).round(2)
print(group_by_year)
highest_sales_idx=group_by_year['total_sales'].idxmax()
print("the highest sales is:",highest_sales_idx)
highest_profit_margin=group_by_year['total_profits'].idxmax()
print("The highest profit margin is:",highest_profit_margin)
```

```python
# # Task 10 — Ship Mode Analysis
# Group by ship_mode
# Find total orders, total sales, average shipping cost per mode
ship_mode_analysis=df.groupby('ship_mode').agg(
    total_orders=('order_id','count'),
    total_sales=('sales','sum'),
    average_shipping_cost=('shipping_cost','mean')
).round(2)

print(ship_mode_analysis)
print("The most used ship mode is:",ship_mode_analysis['total_orders'].idxmax())
print("The ship mode which raises highest sales is:",ship_mode_analysis['total_sales'].idxmax())
```

```python
# Task 11 — Sub-category deep dive
#Group by sub_category
#Find total profit per sub-category

sub_category_analysis=df.groupby('sub_category').agg(
    total_profit=('profit','sum')
).round(2)
print(sub_category_analysis)
print("-------------------------------------------------------------------------")
print("The top 3 most profitable sub-categories:\n",sub_category_analysis.sort_values(by='total_profit',ascending=False).head(3))
print("-------------------------------------------------------------------------")
print("The bottom 3 loss-making sub-categories:\n",sub_category_analysis.sort_values(by='total_profit',ascending=False).tail(3))
```

```python
# # Task 12 — Customer Segmentation using NumPy
# Calculate total spend per customer using groupby on customer_name
# Convert to NumPy array
total_spend_analysis=df.groupby('customer_name').agg(
    total_spends=('sales','sum')
).round(2)
print(total_spend_analysis)
print("-------------------------------------------------------------------------")
per_25=np.percentile(total_spend_analysis,25)
per_75=np.percentile(total_spend_analysis,75)
high_val=np.sum(total_spend_analysis>per_75)
mid_val=np.sum(total_spend_analysis>=per_25) & (total_spend_analysis<=per_75)
low_val=np.sum(total_spend_analysis<per_25)
print(f"75th percentile threshold : ${per_75:,.2f}")
print(f"25th percentile threshold : ${per_25:,.2f}")
print(f"\nHigh value customers : {high_val}")
print(f"Mid  value customers : {mid_val}")
print(f"Low  value customers : {low_val}")

print("\nTop 5 high value customers:")
print(total_spend_analysis.sort_values(by='total_spends',ascending=False).head(5))
```

```python
# Task 13 — Market and Region Analysis

# Group by market and find total sales
# Group by region and find total profit
market_analysis= df.groupby('market')['sales'].sum().sort_values(ascending=False)

region_analysis=df.groupby('region')['profit'].sum().sort_values(ascending=False)
print(market_analysis)
print(region_analysis)
print("The market which contributes most to global sales is:",market_analysis.idxmax())
print("The region which is most profitable is:",region_analysis.idxmax())
```

```python
# Task 14 — NumPy stats on Sales column
# Using only NumPy on sales column:

# Mean, median, std, min, max
print("the average sales are:",df['sales'].mean().round(2))
print("the median sales are:",df['sales'].median())
print(f'the std: {(df['sales'].std()):.2f}')
print("the max sales are:",np.max(df['sales']))
print("the min sales are:",df['sales'].min())

# Find outlier orders using IQR method (same as COVID project)
sales=df['sales'].values
Q1    = np.percentile(sales, 25)
Q3    = np.percentile(sales, 75)
IQR   = Q3 - Q1
upper = Q3 + 1.5 * IQR

outliers = df[df['sales'] > upper]
print(f"\nOutlier threshold   : ${upper:,.2f}")
print(f"Number of outliers  : {len(outliers)}")
print(f"Highest single order: ${np.max(sales):,.2f}")
```
